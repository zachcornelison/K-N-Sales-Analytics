```python
# Import new csv file containing proper coordinates ran through geocoder
missing_addresses = pd.read_csv('missing_addresses.csv', index_col='CustomerID')
missing_addresses.head(3)
```

```python
# Full outer join to merge location data with coordinates on CustomerID key value
df_addresses = pd.merge(df_customers_address, missing_addresses, on='CustomerID', how='outer')
```

```python
df_addresses.head(3)
```

```python
# drop second full address column and rename to remove suffix
df_addresses = df_addresses.drop(['Full Address_y'], axis=1)
df_addresses = df_addresses.rename(columns={"Full Address_x": "Full Address"})
df_addresses.head(3)
```

```python
# ensure shape matches up (result is 1 less for address which coordinates could not be derived)
df_addresses.shape
```

```python
df_addresses.dtypes
```

```python
#generate a folium map to view locations of customers
def generateBaseMap(default_location=[35.507996, -97.529520], default_zoom_start=11):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
customers = df_addresses.dropna()
```

```python
#create an instance of the map
basemap = generateBaseMap()
```

```python
# render map
customers.apply(lambda row: folium.Circle(location=[row['Latitude'], row['Longitude']],
                                    radius=100, color='Blue', fill_color='Blue').add_to(basemap), axis=1)

basemap
```
