```python
# deleting columns that will not be relevant to this analysis
df_customers = df_customers.drop(['Work Phone','Home Phone','Date1', 'Text1','ConsignCode','BirthMonth','StoreID','CheckFee',
                                  'Num1','Email','ActiveInd','Return','CustNotes','CheckAmt','UseMRW','Locked'], axis=1)
df_customers.head()
```

```python
# drop the first two rows which are just a filler row and the store's info
df_customers = df_customers.drop([0,1])
df_customers.head()
```

```python
# change 'Postal Code' from a float to a string
df_customers['Postal Code'] = df_customers['Postal Code'].astype(str).str[:-2]
df_customers.head()
```
